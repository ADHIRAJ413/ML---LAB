```python
import numpy as np
import matplotlib.pyplot as plt

# Given points (from previous context)
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Calculate means
x_mean = np.mean(x)
y_mean = np.mean(y)

# Calculate sums for the slope (m) and intercept (c) using direct formulas
# m = sum((x_i - x_mean) * (y_i - y_mean)) / sum((x_i - x_mean)^2)
# c = y_mean - m * x_mean

numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean)**2)

m_formula = numerator / denominator
c_formula = y_mean - m_formula * x_mean

print(f"Calculated Slope (m): {m_formula:.4f}")
print(f"Calculated Intercept (c): {c_formula:.4f}")

# Plot the original points
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'o', label='Original points', markersize=8)

# Plot the fitted line using the formula-derived parameters
plt.plot(x, m_formula*x + c_formula, 'b', label=f'Fitted line (Formula Method): y = {m_formula:.2f}x + {c_formula:.2f}')

# Add labels, title, and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Linear Regression using Direct Statistical Formulas')
plt.legend()
plt.grid(True)
plt.show()
```

This code demonstrates the calculation of the linear regression line using direct statistical formulas for the slope (`m`) and y-intercept (`c`). As you can see, this method yields the same result as the `np.linalg.lstsq` method for simple linear regression.

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt

# Original points (as used in previous cells)
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# --- Linear Regression Model Metrics ---
# Parameters from the least squares fit (from previous run)
m_linear = 0.60
c_linear = 2.20
y_pred_linear = m_linear * x + c_linear

print('--- Linear Regression Metrics ---')
mse_linear = mean_squared_error(y, y_pred_linear)
mae_linear = mean_absolute_error(y, y_pred_linear)
rmse_linear = np.sqrt(mse_linear)
r2_linear = r2_score(y, y_pred_linear)

# Adjusted R^2 for linear model (p=1, for 1 independent variable 'x')
n = len(y)
p_linear = 1 # Number of independent variables
adj_r2_linear = 1 - (1 - r2_linear) * (n - 1) / (n - p_linear - 1)

print(f"MSE: {mse_linear:.4f}")
print(f"MAE: {mae_linear:.4f}")
print(f"RMSE: {rmse_linear:.4f}")
print(f"R^2: {r2_linear:.4f}")
print(f"Adjusted R^2: {adj_r2_linear:.4f}")
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Given data
x1 = np.array([1, 1, 2])
x2 = np.array([1, 2, 2])
y = np.array([6, 8, 9])

# Construct the design matrix X
# Add a column of ones for the intercept term (beta_0)
X_design = np.vstack([np.ones(len(x1)), x1, x2]).T

# Display the design matrix
print("Design Matrix (X_design):")
print(X_design)

# Calculate (X^T * X)
XTX = X_design.T @ X_design
print("\n(X^T * X):")
print(XTX)

# Calculate the inverse of (X^T * X)
XTX_inv = np.linalg.inv(XTX)
print("\n(X^T * X)^-1:")
print(XTX_inv)

# Calculate (X^T * Y)
XT_y = X_design.T @ y
print("\n(X^T * Y):")
print(XT_y)

# Calculate the regression coefficients (B = (X^T * X)^-1 * X^T * Y)
coefficients = XTX_inv @ XT_y

# Extract coefficients
intercept = coefficients[0]
beta_x1 = coefficients[1]
beta_x2 = coefficients[2]

print("\nRegression Coefficients:")
print(f"  Intercept (beta_0): {intercept:.4f}")
print(f"  Coefficient for x1 (beta_1): {beta_x1:.4f}")
print(f"  Coefficient for x2 (beta_2): {beta_x2:.4f}")

# Predict y values
y_pred = X_design @ coefficients
print("\nPredicted y values:")
print(y_pred.round(4))

# --- Calculate Performance Metrics ---
print("\n--- Performance Metrics ---")
mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)

# Adjusted R^2
n = len(y) # Number of observations
p = X_design.shape[1] - 1 # Number of predictors (x1, x2) - excluding intercept

print(f"MSE: {mse:.4f}")
print(f"MAE: {mae:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R^2: {r2:.4f}")

# Handle the case where Adjusted R^2 is undefined
if (n - p - 1) == 0:
    print("Adjusted R^2: Undefined (n - p - 1 = 0, model perfectly fits data)")
else:
    adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
    print(f"Adjusted R^2: {adj_r2:.4f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Given data (from previous execution)
x1 = np.array([1, 1, 2])
x2 = np.array([1, 2, 2])
y = np.array([6, 8, 9])

# Regression coefficients (from previous execution)
intercept = 3.00
beta_x1 = 1.00
beta_x2 = 2.00

# Predicted y values
y_pred = intercept + beta_x1 * x1 + beta_x2 * x2

# Create the plot
plt.figure(figsize=(8, 8))
plt.scatter(y, y_pred, color='blue', s=150, label='Actual vs. Predicted Points')

# Add a perfect prediction line (y = x) for reference
min_val = min(np.min(y), np.min(y_pred))
max_val = max(np.max(y), np.max(y_pred))
plt.plot([min_val - 1, max_val + 1], [min_val - 1, max_val + 1], 'r--', label='Perfect Prediction (y=x)')

plt.title('Actual vs. Predicted Values for Multiple Linear Regression')
plt.xlabel('Actual Values (y)')
plt.ylabel('Predicted Values (ŷ)')
plt.legend()
plt.grid(True)
plt.axis('equal') # Ensure equal scaling for x and y axes
plt.show()
```

This plot visually confirms the perfect fit of the multiple linear regression model. All the data points lie exactly on the 'Perfect Prediction' line (where Actual = Predicted), which is consistent with the R^2 value of 1.0 and MSE/MAE/RMSE of 0. This indicates that the model perfectly captures the relationship between the independent variables (`x1`, `x2`) and the dependent variable (`y`) for the given training data.

The code above performs multiple linear regression using the **Normal Equation method**, which is a matrix factorization technique. Here's a breakdown of the output:

*   **Design Matrix (X_design)**: This matrix includes a column of ones for the intercept term, followed by the `x1` and `x2` values.
*   **Matrix Operations (X^T * X), (X^T * X)^-1, (X^T * Y)**: These are the intermediate calculations required by the normal equation `B = (X^T * X)^-1 * X^T * Y`.
*   **Regression Coefficients**: These are the values for `beta_0` (intercept), `beta_1` (coefficient for x1), and `beta_2` (coefficient for x2) that define your regression line `y = beta_0 + beta_1*x1 + beta_2*x2`.
*   **Predicted y values**: These are the `y` values estimated by the regression model.
*   **Performance Metrics**: These metrics (MSE, MAE, RMSE, R^2, Adjusted R^2) evaluate how well the regression model fits the given data. Given only 3 data points and 2 predictors, the R^2 and Adjusted R^2 are very high, indicating a perfect or near-perfect fit to the training data, which is expected with limited data points.
